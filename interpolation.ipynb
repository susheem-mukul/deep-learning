```
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```
from google.colab import files
from io import BytesIO
from PIL import Image

uploaded = files.upload()
im = Image.open(BytesIO(uploaded['apple.jpg']))
```

```
plt.imshow(im)
```

```
m=np.asarray(im)
```

```
out=cv2.resize(m,None,fx=3,fy=1,interpolation=cv2.INTER_LINEAR)
```

```
plt.imshow(out)
```
